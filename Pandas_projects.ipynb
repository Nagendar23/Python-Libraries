# Project 1: Weekly Expense Tracker

What you practice: DataFrames, column operations, filtering, simple stats

## Idea:
Track daily expenses for one week and analyze your spending.

### Tasks:
Create a table with days and expenses, then:    
calculate total and average spending    
find the most expensive day     
list days where spending was above average

```python
# week expense tracker
import pandas as pd
data = {
    'Days':['sun','mon','tue','wedn','thur','fri','sat'],
    'Expenses':[1000,250,320,280,390,290,500]
}

df = pd.DataFrame(data)
print('the created expenses data is :')
print(df)

# calculate total and average spending
total = df['Expenses'].sum()
print('total expense is ',total)
avg = df['Expenses'].mean()
print('the average spending of week is :',avg)
print()

# find the most expensive day
expensive_day = df.loc[df['Expenses'].idxmax()]
print(expensive_day)
print()

# list days where spending was above average
above_avg = df[df['Expenses']> avg]
print('the above spending than avg days are : ',above_avg)
```

# 3) Weather Data Explorer

<!-- Time series without terror. -->
### Using daily weather data (date, temperature, rainfall):
Find hottest and coldest days       
Compute monthly averages        
Detect trends over time

### The goal

### You have daily weather records:
a date      
a temperature           
a rainfall amount

#### You want to:
Find the hottest and coldest days       
Compute monthly average temperature and rainfall        
See trends over time (is it warming? wetter?)

```python

import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('weather_data.csv')
print(df)

#better to convert into actual datatime format
df['date'] = pd.to_datetime(df['date'])
print()

# find the hottest and coldest days
hottest_day = df.loc[df['temperature_c'].idxmax()]
print('hottest day')
print(hottest_day)
print()
coldest_day = df.loc[df['temperature_c'].idxmin()]
print('coldest_day')
print(coldest_day)
print()

# compute monthly average temperature and rainfall

avg_temp = df['temperature_c'].mean()
print('The average temperature is :',avg_temp)
avg_rainfall = df['rainfall_mm'].mean()
print('the average rainfall mm is :',avg_rainfall)
print()

# observing trends overtime (is it warming or wetter)  simple visual trend

df['temperature_c'].plot(title='Daily temperature over time')
plt.ylabel('Temperature in C')
plt.show()
```

# 4) Sales Performance Dashboard (Mini Version)

Business intelligence in miniature.     
With sales records (date, product, region, revenue):        
Total revenue per product       
Best-performing region      
Monthly sales growth (Are sales growing or shrinking month by month? ) can use pivot tables

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('sales_data.csv')
#better to change into actual date time format
df['date'] = pd.to_datetime(df['date'])
# print(df) 
print()

#total revenue per product
ttl_rev_per_prod   = df.groupby('product')['revenue'].sum()
print('total revenue per product :')
print(ttl_rev_per_prod)
print()

#best performing region
revenue_by_region = df.groupby('region')['revenue'].sum()
print('revenue acc to region :',revenue_by_region)
bst_perform_reg = revenue_by_region.idxmax()
print("The best performing refion is :",bst_perform_reg)
print()
# monthly sales growth

# here we need to make the date as index
df = df.set_index('date')
# df['revenue'].plot(title='Sales over month')
# plt.show()
monthly_sales = df.resample('ME')['revenue'].sum()  # here we will consider acc to month and total of that month
print(monthly_sales)
monthly_growth = monthly_sales.pct_change()*100      #tcalculates how much sales changed compared to the previous month,
print(monthly_growth)
#visualizng it
monthly_sales.plot(title='monthly sales revenue',ylabel='revenue')
plt.show()


```

# 5) Data Cleaning Playground

Start with a deliberately ugly dataset:

Inconsistent column names       
Missing values      
Duplicate rows      
Mixed data types

1. Inconsistent column names        
Columns might look like:        
Customer Name       
customer_name       
CustomerName        

2. Missing values       
You’ll see:     
empty cells     
NaN     
missing ages, prices, or categories         
Missing data is not an error — it’s information about how data was collected.

3. Duplicate rows       
Same record appears twice because:      
someone clicked “submit” twice      
a system retried a failed request       
files were merged badly         
Duplicates silently corrupt totals if you ignore them.

4. Mixed data types     
Examples:       
numbers stored as strings ("300" instead of 300)        
text mixed with numbers ("N/A", "unknown")      
Pandas will refuse to do math until you clean this up.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('dirty_sales_dataset.csv')
# print(df.info)
# print(df.head())

#1. Inconsistent column names => fix it
# Columns might look like:
# Customer Name
# customer_name
# CustomerName
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ','_')
)
df['customer_name'] = df['customer_name'].str.strip().str.title()   # Capitalize the name
# print(df)

# 2. Missing values       
# You’ll see:     
# empty cells     
# NaN     
# missing ages, prices, or categories         
# Missing data is not an error — it’s information about how data was collected.
df['sale_amount'] = pd.to_numeric(df['sale_amount'],errors='coerce') 
df['age'] = pd.to_numeric(df['age'],errors='coerce')       #converts N/A into NaN

# handling missing values
df['sale_amount'] = df['sale_amount'].fillna(df['sale_amount'].median())  #we can also use mean but median give better results
df['age'] = df['age'].fillna(df['age'].median())

#3. Remove Duplicate rows       
df = df.drop_duplicates()

print(df)
```

# 6) Simple CSV → Report Generator

#### Turn raw data into something a human might read.

Load a CSV, compute key statistics, and export:     
A cleaned CSV       
A summary table         
A basic chart

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('report_gen.csv')
print(df.info())
print()


#cleaning the data
# we have the date so convert into normal format
df['date'] = pd.to_datetime(df['date'])
# then make the unknown or N/A into NaN
df['revenue'] = pd.to_numeric(df['revenue'],errors='coerce')
#fill the missing values
df['revenue'] = df['revenue'].fillna(df['revenue'].median())


#summary table ( we can create revenue by product, region)

print('creating summary table by product')
summary_product = df.groupby('product')['revenue'].agg(      #with help of the agg we can specify multiple calc values along with column names 
    total_revenue="sum",
    avg_revenue="mean",
    transaction_count="count"
)
print(summary_product)
print()

# creating summary table bu region
print('creating summary table bu region')
summary_region=df.groupby('region')['revenue'].agg(
    total_revenue="sum",
    avg_revenue="mean",
    transaction_count="count"
)
print(summary_region)

#creating a basic chart acc to prod
summary_product['total_revenue'].plot(
    kind='bar',
    title='total revenue by product',
    ylabel='revenue'
)
plt.show()

#creating a basic chart acc to prod
summary_region['total_revenue'].plot(
    kind='bar',
    title='total revenue acc to region',
    ylabel='revenue'
)
plt.show()
```
