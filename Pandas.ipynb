### Series
A Pandas Series is like a column in a table.

It basically prionts the given values with index no.s

use pd.Series(x)

```python
import pandas as pd
a=[1,3,6]
res=pd.Series(a)
print(res)
```

### Labels
If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.

This label can be used to access a specified value.

```python
import pandas as pd
a=[1,3,6]
res=pd.Series(a)
#label
print(res[1])
```

#### Create Labels
With the index argument, you can name your own labels.

```python
import pandas as pd
b=[1,3,6]
res=pd.Series(a,index=['x','y','z'])
print(res)
#we can access using the ddefined index 
print(res['y'])
```

```python
# Key/Value Objects as Series
# You can also use a key/value object, like a dictionary, when creating a Series.

import pandas as np
screentime={'phone':5, 'laptop':7,'book':1, 'walking':1}
res = pd.Series(screentime)
print(res)
```

```python
# To select only some of the items in the dictionary, use the index argument and specify only the items you want to include in the Series.
import pandas as np
screentime={'phone':5, 'laptop':7,'book':1, 'walking':1}
res = pd.Series(screentime,index=['laptop','book'])
print(res)
```

### DataFrames
Data sets in Pandas are usually multi-dimensional tables, called DataFrames.

A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

Basically it is used to create table like strcture data

Series is like a column, a DataFrame is the whole table.

```python
import pandas as pd
data = {
'calories':[420,380,512],
'duration':[50,45,52]
}
res=pd.DataFrame(data)
print(res)
```

```python
# Pandas use the loc attribute to return one or more specified row(s)

print(res.loc[1])   # give all rows of index 1 column
print()
print()
print(res.loc[[0,1]])   # give all rows of index 0,1 column
```

```python
# Named Indexes
# With the index argument, you can name your own indexes.
import pandas as pd
data = {
'calories':[420,380,512],
'duration':[50,45,52]
}
res=pd.DataFrame(data,index=['day1','day2','day3'])
print(res)
```

### Analyzing DataFrames
Viewing the Data    
One of the most used method for getting a quick overview of the DataFrame, is the head() method.

The head() method returns the headers and a specified number of rows, starting from the top.

```python
import pandas as pd
load_data = pd.read_csv('pdData.csv')
print(load_data.head(10))  #gives first 10 rows
```

```python
#if we don't mention the value in head() , it will print first 5 rows
import pandas as pd
df = pd.read_csv('pdData.csv')
print(df.head())
```

```python
# There is also a tail() method for viewing the last rows of the DataFrame.

import pandas as pd
df = pd.read_csv('pdData.csv')
print(df.tail())       # gives last 5 rows
```
